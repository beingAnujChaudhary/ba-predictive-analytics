```python
# Calculate total eligible passengers
df['TOTAL_ELIGIBLE_PAX'] = df['TIER1_ELIGIBLE_PAX'] + df['TIER2_ELIGIBLE_PAX'] + df['TIER3_ELIGIBLE_PAX']
# Calculate total seats
df['TOTAL_SEATS'] = df['FIRST_CLASS_SEATS'] + df['BUSINESS_CLASS_SEATS'] + df['ECONOMY_SEATS']
print(f"\nTotal eligible passengers across all flights: {df['TOTAL_ELIGIBLE_PAX'].sum():,}")
print(f"  - Tier 1 eligible: {df['TIER1_ELIGIBLE_PAX'].sum():,}")
print(f"  - Tier 2 eligible: {df['TIER2_ELIGIBLE_PAX'].sum():,}")
print(f"  - Tier 3 eligible: {df['TIER3_ELIGIBLE_PAX'].sum():,}")
```

```python
# ============================================================================
# STEP 4: CREATE FLIGHT GROUPINGS
# ============================================================================
print("\n" + "=" * 80)
print("STEP 4: Creating Flight Groupings")
print("=" * 80)
```

```python
# Strategy: Group flights by meaningful categories
# We'll use: HAUL + TIME_OF_DAY + ARRIVAL_REGION

# Create time period categories
def categorize_time_of_day(time_str):
    """Categorize flight time into broader periods"""
    if time_str in ['Morning']:
        return 'Morning (06:00-11:59)'
    elif time_str in ['Lunchtime']:
        return 'Midday (12:00-16:59)'
    elif time_str in ['Afternoon']:
        return 'Afternoon (12:00-16:59)'
    elif time_str in ['Evening']:
        return 'Evening (17:00-23:59)'
    else:
        return 'Other'

df['TIME_PERIOD'] = df['TIME_OF_DAY'].apply(categorize_time_of_day)

# Create route category combining haul and region
df['ROUTE_CATEGORY'] = df['HAUL'] + ' - ' + df['ARRIVAL_REGION']

# Create comprehensive category
df['FLIGHT_CATEGORY'] = df['HAUL'] + '_' + df['TIME_PERIOD'].str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace(':', '')

print("\nFlight grouping strategy:")
print("  1. By HAUL (SHORT vs LONG)")
print("  2. By TIME_PERIOD (Morning, Midday, Afternoon, Evening)")
print("  3. By ARRIVAL_REGION (Europe, North America, Asia, Middle East)")
print("  4. Combined: HAUL + TIME_PERIOD")
```

```python
print("\n" + "-" * 80)
print("Distribution by Flight Category (HAUL + TIME_PERIOD):")
print("-" * 80)
print(df['FLIGHT_CATEGORY'].value_counts())

print("\n" + "-" * 80)
print("Distribution by Route Category (HAUL + REGION):")
print("-" * 80)
print(df['ROUTE_CATEGORY'].value_counts())
```

```python
# ============================================================================
# STEP 5: CALCULATE ELIGIBILITY PERCENTAGES BY CATEGORY
# ============================================================================
print("\n" + "=" * 80)
print("STEP 5: Calculating Eligibility Percentages by Category")
print("=" * 80)

# Function to calculate eligibility percentages
def calculate_eligibility_percentages(group):
    """Calculate the percentage of passengers eligible for each tier"""
    total_tier1 = group['TIER1_ELIGIBLE_PAX'].sum()
    total_tier2 = group['TIER2_ELIGIBLE_PAX'].sum()
    total_tier3 = group['TIER3_ELIGIBLE_PAX'].sum()
    total_eligible = total_tier1 + total_tier2 + total_tier3
    
    # Calculate total passengers (sum of all eligible + estimate of non-eligible)
    # We'll use total seats as a proxy for total passengers
    total_seats = group['TOTAL_SEATS'].sum()
    
    # Calculate percentages based on total eligible passengers
    if total_eligible > 0:
        tier1_pct = (total_tier1 / total_eligible) * 100
        tier2_pct = (total_tier2 / total_eligible) * 100
        tier3_pct = (total_tier3 / total_eligible) * 100
    else:
        tier1_pct = tier2_pct = tier3_pct = 0
    
    return pd.Series({
        'Total_Flights': len(group),
        'Total_Seats': total_seats,
        'Tier1_Eligible': total_tier1,
        'Tier2_Eligible': total_tier2,
        'Tier3_Eligible': total_tier3,
        'Total_Eligible': total_eligible,
        'Tier1_Percent': round(tier1_pct, 1),
        'Tier2_Percent': round(tier2_pct, 1),
        'Tier3_Percent': round(tier3_pct, 1)
    })

# Calculate percentages by different groupings
print("\n" + "-" * 80)
print("A. Eligibility by HAUL + TIME_PERIOD:")
print("-" * 80)

eligibility_by_category = df.groupby('FLIGHT_CATEGORY').apply(calculate_eligibility_percentages)
print(eligibility_by_category)

print("\n" + "-" * 80)
print("B. Eligibility by HAUL + REGION:")
print("-" * 80)

eligibility_by_route = df.groupby('ROUTE_CATEGORY').apply(calculate_eligibility_percentages)
print(eligibility_by_route)

print("\n" + "-" * 80)
print("C. Eligibility by TIME_OF_DAY only:")
print("-" * 80)

eligibility_by_time = df.groupby('TIME_OF_DAY').apply(calculate_eligibility_percentages)
print(eligibility_by_time)

print("\n" + "-" * 80)
print("D. Eligibility by HAUL only:")
print("-" * 80)

eligibility_by_haul = df.groupby('HAUL').apply(calculate_eligibility_percentages)
print(eligibility_by_haul)
```

```python
# ============================================================================
# STEP 6: CREATE THE LOOKUP TABLE
# ============================================================================
print("\n" + "=" * 80)
print("STEP 6: Creating the Reusable Lookup Table")
print("=" * 80)

# Based on analysis, we'll create a lookup table using HAUL + TIME_PERIOD
# This provides a good balance between granularity and generalizability

# Create the final lookup table
lookup_table = df.groupby('FLIGHT_CATEGORY').apply(calculate_eligibility_percentages).reset_index()

# Rename columns for clarity
lookup_table = lookup_table.rename(columns={
    'FLIGHT_CATEGORY': 'Flight_Category',
    'Tier1_Percent': 'Tier_1_Percent',
    'Tier2_Percent': 'Tier_2_Percent',
    'Tier3_Percent': 'Tier_3_Percent',
    'Total_Flights': 'Number_of_Flights',
    'Total_Seats': 'Total_Seats',
    'Tier1_Eligible': 'Tier_1_Eligible_Passengers',
    'Tier2_Eligible': 'Tier_2_Eligible_Passengers',
    'Tier3_Eligible': 'Tier_3_Eligible_Passengers',
    'Total_Eligible': 'Total_Eligible_Passengers'
})

# Reorder columns
lookup_table = lookup_table[[
    'Flight_Category',
    'Number_of_Flights',
    'Total_Seats',
    'Tier_1_Eligible_Passengers',
    'Tier_2_Eligible_Passengers',
    'Tier_3_Eligible_Passengers',
    'Total_Eligible_Passengers',
    'Tier_1_Percent',
    'Tier_2_Percent',
    'Tier_3_Percent'
]]

print("\n" + "=" * 80)
print("FINAL LOOKUP TABLE - Lounge Eligibility by Flight Category")
print("=" * 80)
print(lookup_table.to_string(index=False))

# Create a simpler version for the Excel template
simple_lookup = lookup_table[['Flight_Category', 'Tier_1_Percent', 'Tier_2_Percent', 'Tier_3_Percent']].copy()

print("\n" + "-" * 80)
print("Simplified Lookup Table (for Excel template):")
print("-" * 80)
print(simple_lookup.to_string(index=False))
```

```python
# ============================================================================
# STEP 7: APPLY LOOKUP TABLE TO SAMPLE FLIGHTS
# ============================================================================
print("\n" + "=" * 80)
print("STEP 7: Applying Lookup Table to Sample Flights")
print("=" * 80)

# Select a representative sample of flights
sample_flights = df.head(10).copy()

# Apply the lookup table to estimate lounge demand
def estimate_lounge_demand(flight_row, lookup_df):
    """Estimate lounge demand for a single flight"""
    category = flight_row['FLIGHT_CATEGORY']
    
    # Find the matching category in lookup table
    match = lookup_df[lookup_df['Flight_Category'] == category]
    
    if len(match) > 0:
        tier1_pct = match['Tier_1_Percent'].values[0] / 100
        tier2_pct = match['Tier_2_Percent'].values[0] / 100
        tier3_pct = match['Tier_3_Percent'].values[0] / 100
        
        # Estimate eligible passengers based on total seats
        total_seats = flight_row['TOTAL_SEATS']
        
        estimated_tier1 = round(total_seats * tier1_pct * 0.1)  # 10% load factor assumption
        estimated_tier2 = round(total_seats * tier2_pct * 0.1)
        estimated_tier3 = round(total_seats * tier3_pct * 0.1)
        
        return pd.Series({
            'Estimated_Tier1': estimated_tier1,
            'Estimated_Tier2': estimated_tier2,
            'Estimated_Tier3': estimated_tier3,
            'Total_Estimated': estimated_tier1 + estimated_tier2 + estimated_tier3
        })
    else:
        return pd.Series({
            'Estimated_Tier1': 0,
            'Estimated_Tier2': 0,
            'Estimated_Tier3': 0,
            'Total_Estimated': 0
        })

# Apply to sample
sample_estimates = sample_flights.apply(lambda row: estimate_lounge_demand(row, lookup_table), axis=1)
sample_flights = pd.concat([sample_flights, sample_estimates], axis=1)

print("\nSample Flight Analysis with Estimated Lounge Demand:")
print("-" * 80)

display_cols = [
    'FLIGHT_NO', 'FLIGHT_TIME', 'TIME_OF_DAY', 'ARRIVAL_STATION_CD',
    'HAUL', 'TOTAL_SEATS', 'Estimated_Tier1', 'Estimated_Tier2',
    'Estimated_Tier3', 'Total_Estimated'
]

print(sample_flights[display_cols].to_string(index=False))
```

```python
# ============================================================================
# STEP 8: CREATE OUTPUT FILES
# ============================================================================
print("\n" + "=" * 80)
print("STEP 8: Creating Output Files")
print("=" * 80)

# Create outputs directory
outputs_dir = Path('outputs')
outputs_dir.mkdir(exist_ok=True)

tables_dir = outputs_dir / 'tables'
tables_dir.mkdir(exist_ok=True)

# Save the lookup table to Excel
lookup_file = tables_dir / 'lounge_eligibility_lookup_table.xlsx'

with pd.ExcelWriter(lookup_file, engine='openpyxl') as writer:
    # Sheet 1: Main lookup table
    lookup_table.to_excel(writer, sheet_name='Lookup_Table', index=False)
    
    # Sheet 2: Simplified version
    simple_lookup.to_excel(writer, sheet_name='Simple_Lookup', index=False)
    
    # Sheet 3: Summary statistics
    summary_stats = pd.DataFrame({
        'Metric': [
            'Total Flights Analyzed',
            'Total Seats',
            'Total Tier 1 Eligible',
            'Total Tier 2 Eligible',
            'Total Tier 3 Eligible',
            'Overall Tier 1 %',
            'Overall Tier 2 %',
            'Overall Tier 3 %'
        ],
        'Value': [
            len(df),
            df['TOTAL_SEATS'].sum(),
            df['TIER1_ELIGIBLE_PAX'].sum(),
            df['TIER2_ELIGIBLE_PAX'].sum(),
            df['TIER3_ELIGIBLE_PAX'].sum(),
            round((df['TIER1_ELIGIBLE_PAX'].sum() / df['TOTAL_ELIGIBLE_PAX'].sum()) * 100, 1),
            round((df['TIER2_ELIGIBLE_PAX'].sum() / df['TOTAL_ELIGIBLE_PAX'].sum()) * 100, 1),
            round((df['TIER3_ELIGIBLE_PAX'].sum() / df['TOTAL_ELIGIBLE_PAX'].sum()) * 100, 1)
        ]
    })
    summary_stats.to_excel(writer, sheet_name='Summary_Statistics', index=False)
    
    # Sheet 4: Sample application
    sample_flights[display_cols].to_excel(writer, sheet_name='Sample_Application', index=False)

print(f"\n✓ Lookup table saved to: {lookup_file}")
```
